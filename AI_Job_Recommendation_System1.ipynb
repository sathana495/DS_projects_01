```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/dataset/jobs.csv')
df.head(3)
```

```python
# Read dataset - using the already loaded DataFrame
jobs = df.to_dict('records')


user_skills = input("Enter your skills (comma separated): ").lower().split(",")

user_skills = [skill.strip() for skill in user_skills]

results = []

# Compare skills
for job in jobs:

    job_role = job["Job Role"]

    # Ensure job_skills is handled correctly, even if it's not a string or missing
    job_skills_str = job.get("Skills", "") # Get 'Skills' column, default to empty string if missing
    job_skills = job_skills_str.lower().split(",")

    matched_skills = []

    missing_skills = []

    for skill in job_skills:

        skill = skill.strip()

        if skill and skill in user_skills: # Added 'if skill' to handle empty strings from split
            matched_skills.append(skill)
        elif skill: # Only add non-empty skills to missing
            missing_skills.append(skill)

    # Handle cases where job_skills might be empty to avoid division by zero
    score = (len(matched_skills) / len(job_skills)) * 100 if job_skills else 0

    results.append({
        "role": job_role,
        "score": score,
        "missing": missing_skills
    })

# Sort by score
results.sort(key=lambda x: x["score"], reverse=True)

print("\n========== TOP 3 JOB RECOMMENDATIONS ==========\n")

for i in range(3):
    if i < len(results): # Ensure we don't try to access out of bounds if there are less than 3 results
        print(f"{i+1}. {results[i]['role']}")
        print(f"Match Score: {results[i]['score']:.2f}%")

        print("Missing Skills:")

        if results[i]["missing"]: # Only print if there are missing skills
            for skill in results[i]["missing"]:
                print(" -", skill)
        else:
            print(" - None")

        print()

# Save result
with open("result.txt", "w") as file:

    file.write("TOP JOB RECOMMENDATIONS\n\n")

    for i in range(3):
        if i < len(results):
            file.write(f"{i+1}. {results[i]['role']}\n")
            file.write(f"Score: {results[i]['score']:.2f}%\n\n")

print("Results saved in result.txt")
```
